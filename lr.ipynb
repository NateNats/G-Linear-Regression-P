```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
```

```python
X, y = make_regression(n_samples=100, n_features=1, random_state=24, noise=20)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, random_state=24)
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(X, y)
plt.show()
```

```python
from linreg import Linear_Regression

lr = Linear_Regression()
lr.train(X_train, y_train)
```

```python
predict_val = lr.predict(X_test)
print(predict_val)
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression

lr_sk = Linear_Regression()
lr_sk.train(X_train, y_train)
```

```python
pre_val_sk = lr_sk.predict(X_test)
print(pre_val_sk)
```

```python
# checking output of lr scikit learn and lr from scratch
print(predict_val == pre_val_sk)
```

```python
# show x_test and y_pred from scratch model
plt.scatter(X_test, y_test)
plt.title("lr from scratch predicton output")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```
